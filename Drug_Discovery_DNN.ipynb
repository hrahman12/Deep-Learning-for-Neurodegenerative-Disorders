```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('Glycine_06_bioactivity_data_3class_pIC50_pubchem_fp.csv')
```

```python
X = df.drop('pIC50', axis=1)
X
```

```python
Y = df.pIC50
Y
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))
X = selection.fit_transform(X)
```

```python
X.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
```

```python
import keras
from keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(128, kernel_initializer='normal', input_shape=(212,), activation='relu'))

model.add(Dense(256,kernel_initializer='normal', activation='relu'))
model.add(Dense(256,kernel_initializer='normal', activation='relu'))
model.add(Dense(256,kernel_initializer='normal', activation='relu'))

model.add(Dense(1,kernel_initializer='normal', activation='linear'))

model.compile(loss='mean_squared_error', optimizer='adam',metrics=['mean_absolute_error'])
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
checkpoint_name = 'Weights-{epoch:03d}--{val_loss:.5f}.hdf5'
checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
callbacks_list = [checkpoint]
```

```python
model.fit(X_train, Y_train, epochs=250, batch_size=32, validation_data=(X_test,Y_test), callbacks=callbacks_list)
```

```python
# Load wights file of the best model :
weights_file = r'C://Users/Rahma/QML FOR MS/Weights-037--1.68439.hdf5' # choose the best checkpoint
model.load_weights(weights_file) # load it
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

Y_test_pred = model.predict(X_test)

r2_score = r2_score(Y_test, Y_test_pred)
print("The accuracy of our model is {}%".format(round(r2_score, 2) *100))
```
